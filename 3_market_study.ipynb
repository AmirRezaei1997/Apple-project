```python
# prepare the dataframe
df1=df.loc[~(df['Country']=='United Kingdom')]
df1['Total']=df['UnitPrice']*df['Quantity']
df2=df1.drop_duplicates(['CustomerId','Country'])
###################
df3=df2.groupby('Country').CustomerId.size()
#df3.drop_duplicates('CustomerId')
df4=df1.groupby('Country').Total.sum()
df3.quantile(0.75)
df4.quantile(0.75)
df5=pd.concat([df3, df4.reindex(df3.index)], axis=1)
G1=df5[(df5.CustomerId<=df3.quantile(0.75)) & (df5.Total<=df4.quantile(0.75))]
G2=df5[(df5.CustomerId>df3.quantile(0.75)) & (df5.Total<df4.quantile(0.75))]
G3=df5[(df5.CustomerId<df3.quantile(0.75)) & (df5.Total>df4.quantile(0.75))]
G4=df5[(df5.CustomerId>df3.quantile(0.75)) & (df5.Total>df4.quantile(0.75))]
G5=G4.iloc[[1,3,4]]

df3
```

```python
# plot

plt.style.use("fivethirtyeight")
fig , ax = plt.subplots(figsize=(10,10))
ax.scatter(G4.Total,G4.CustomerId ,color='red',s=20 ,label='Highest customer & revenue')
#ax.scatter(G5.Total,G5.CustomerId ,color='red',s=20 ,label='Highest customer & revenue')
ax.scatter(G2.Total,G2.CustomerId ,color='blue',s=20 ,label='High customer & low revenue')
ax.scatter(G3.Total,G3.CustomerId ,color='green',s=20 ,label='High revenue & low customer')
ax.scatter(G1.Total,G1.CustomerId,color='cyan' ,s=20 ,label='Low customer & revenue')
ax.set_title('Market Study' ,fontsize=18)
ax.set_xlabel('Sales amount' ,fontsize=16)
ax.set_ylabel('Number of customers' ,fontsize=16)
ax.legend(frameon=True ,title='Group',title_fontsize=16 ,facecolor='silver' ,fontsize=14)
annotations=["France","Netherlands","Spain"]
for i, label in enumerate(annotations):
    plt.annotate(label, (G5.Total[i], G5.CustomerId[i]+0.5),fontsize=15)

```
