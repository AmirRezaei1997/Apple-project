```python
# import libraries: TODO
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import date
```

```python
# read the preprocessed dataset: TODO
df=pd.read_csv(r'D:\quera\final2\final_project\2_exploration\preprocessed_sales.csv')
df['InvoiceDate']=pd.to_datetime(df['InvoiceDate'])
df['InvoiceDate'] = df['InvoiceDate'].dt.date
date1=df.InvoiceDate.max()
date1
today = date(2010, 12, 10)
```

```python
df['Days']= (today-df['InvoiceDate'])
df['total']=(df.Quantity)*(df.UnitPrice)
df
```

```python
me=df.copy()
le=me.drop_duplicates(subset=['InvoiceNumber'])
#df3=df.groupby('CustomerId').InvoiceNumber.size()
df3=le.groupby('CustomerId').InvoiceNumber.size()
df3
```

```python
df4=df.groupby('CustomerId').total.sum()
df4
```

```python
df1 = df.groupby('CustomerId')['Days'].min()
df1
```

```python
df5=pd.concat([df1,df3, df4.reindex(df3.index)], axis=1)
df5
```

```python
d1=[(df5['Days']<=df5['Days'].quantile(0.25)),
(df5['Days']>df5['Days'].quantile(0.25))&((df5['Days']<=df5['Days'].quantile(0.5))),
(df5['Days']>df5['Days'].quantile(0.5))&((df5['Days']<=df5['Days'].quantile(0.75))),
(df5['Days']>df5['Days'].quantile(0.75))]
d2=['1','2','3','4']
df5['R_quartile']=np.select(d1,d2)
#############################################
f1=[(df5['InvoiceNumber']<=df5['InvoiceNumber'].quantile(0.25)),
(df5['InvoiceNumber']>df5['InvoiceNumber'].quantile(0.25))&((df5['InvoiceNumber']<=df5['InvoiceNumber'].quantile(0.5))),
(df5['InvoiceNumber']>df5['InvoiceNumber'].quantile(0.5))&((df5['InvoiceNumber']<=df5['InvoiceNumber'].quantile(0.75))),
(df5['InvoiceNumber']>df5['InvoiceNumber'].quantile(0.75))]
f2=['1','2','3','4']
df5['F_quartile']=np.select(f1,f2)
###########################################
m1=[(df5['total']<=df5['total'].quantile(0.25)),
(df5['total']>df5['total'].quantile(0.25))&((df5['total']<=df5['total'].quantile(0.5))),
(df5['total']>df5['total'].quantile(0.5))&((df5['total']<=df5['total'].quantile(0.75))),
(df5['total']>df5['total'].quantile(0.75))]
m2=['1','2','3','4']
df5['M_quartile']=np.select(m1,m2)
###########################################
df5['RFM']=df5['R_quartile']+df5['F_quartile']+df5['M_quartile']
###########################################
df5
```
